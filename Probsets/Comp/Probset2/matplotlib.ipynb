### Problem 1

```python
import matplotlib.pyplot as plt
import numpy as np


def mean_var_norm(n):
    draws = np.random.normal(size=(n, n))
    row_mean = np.mean(draws, axis=1)
    return np.var(row_mean)

def variance_rows():
    N = np.arange(100, 1000, 100)
    return np.array([mean_var_norm(n) for n in N])

plt.plot(variance_rows())
plt.show()
```

### Problem 2

```python
def plot_trig():
    x = np.arange(-np.pi, np.pi, 0.01)
    plt.plot(x, np.sin(x), label='$sin(x)$')
    plt.plot(x, np.cos(x), label='$cos(x)$')
    plt.plot(x, np.arctan(x), label='$arctan(x)$')
    plt.legend()

    
plot_trig()
```

### Problem 3

```python
def hyperbole():
    x_l = np.arange(-2, 0.99, 0.01)
    x_r = np.arange(1.01, 6, 0.01)
    f = lambda x: 1 / (x - 1)
    plt.plot(x_l, f(x_l), ls='--', color='magenta')
    plt.plot(x_r, f(x_r), ls='--', color='magenta')
    plt.xlim(-2, 6)
    plt.ylim(-6, 6)
    plt.show()
```

```python
hyperbole()
```

### Problem 4

```python
def plot_trig2():
    
    fig, axes = plt.subplots(2, 2)
    x_grid = np.linspace(0, 2 * np.pi)
    
    axes[0, 0].plot(x_grid, np.sin(x_grid), 'g')
    axes[0, 1].plot(x_grid, np.sin(2 * x_grid), 'r--')
    axes[1, 0].plot(x_grid, 2 * np.sin(x_grid), 'b--')
    axes[1, 1].plot(x_grid, 2 * np.sin(2 * x_grid), 'm:')
    
    titles = ['$sin(x)$', '$sin(2x)$', '$2sin(x)$', '$2sin(2x)$']
    
    for ax, title in zip(axes.flatten(), titles):
        ax.set(xlim=(0, np.pi), ylim=(-2, 2), title=title)
    
    fig.tight_layout()
    fig.suptitle('Trig functions')
    plt.show()
```

```python
plot_trig2()
```

### Problem 5

```python
def plot_fars():

    fars = np.load('FARS.npy')

    fig, axes = plt.subplots(1, 2, figsize=(10, 3))

    axes[0].scatter(fars[:, 1], fars[:, 2], c='k', s=0.5)
    axes[0].axis('equal')

    hours = range(0, 25, 1)
    axes[1].hist(fars[:, 0], bins=hours, edgecolor='k')
    axes[1].set(xlim=(0, 24), xlabel='Hours')

    plt.show()
```

```python
plot_fars()
```

### Problem 6

```python
def plot_mesh():
    x_grid, y_grid = np.linspace(-2 * np.pi, 2 * np.pi), np.linspace(-2 * np.pi, 2 * np.pi)
    x, y = np.meshgrid(x_grid, y_grid)
    g = lambda x, y: (np.sin(x) * np.sin(y)) / (x * y)
    
    fig, axes = plt.subplots(1, 2, figsize=(10, 3))
    axes[0].pcolormesh(x, y, g(x, y), cmap='coolwarm')
    axes[1].contour(x, y, g(x, y), 20, cmap='jet')
    
    for ax in axes:
        ax.set(xlim=(-2 * np.pi, 2 * np.pi), ylim=(-2 * np.pi, 2 * np.pi))
        
    plt.show()
```

```python
plot_mesh()
```
