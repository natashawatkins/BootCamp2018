```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import root
```

## Plot 1

```python
m1 = lambda x, y, c: 5 * x - 4 * y + c
c1 = lambda x: 2/3 * x + 4/3
c2 = lambda x: x/6 - 1/6
c3 = lambda x: 6 - x
```

```python
max_i = 6
min_i = 0

xgrid = np.linspace(min_i, max_i)
ygrid = np.linspace(min_i, max_i)

plt.figure(figsize=(8, 6))

# Plot level sets
for c in np.arange(0, 3, 0.4):
    plt.plot(xgrid, m1(xgrid, ygrid, c), '--k', alpha=0.6)
    
# Plot opposite of feasible set
plt.fill_between(xgrid, min_i, c1(xgrid), color='lightblue')
plt.fill_between(xgrid, min_i, c2(xgrid), color='lightblue')
plt.fill_between(xgrid, c3(xgrid), max_i, color='lightblue')

# Solve for optimum
f = lambda x: 2/3 * x + 4/3 - (6 - x)
x = root(f, 1).x
y = c1(x)
plt.hlines(y, 0, x, linestyle=':', color='red', label='Optimizer')
plt.vlines(x, 0, y, linestyle=':', color='red')

plt.ylim(min_i, max_i)
plt.xlim(min_i, max_i)
plt.xlabel('$x$')
plt.ylabel('$y$')

# Add a legend
plt.fill_between(xgrid, min_i, min_i, color='white', label='Feasible set')
plt.plot(xgrid, -1 * np.ones(len(xgrid)), '--k', alpha=0.6, label='Level sets')
plt.legend(loc='lower right')

plt.savefig('plot1.png')
plt.show()
```

## Plot 2

```python
m1 = lambda x, y, c: -3 * x + y + c
c1 = lambda x: 5 - 1/3 * x
c2 = lambda x: 6 - 2/3 * x
c3 = lambda x: x - 4
```

```python
max_i = 7
min_i = 0

xgrid = np.linspace(min_i, max_i)
ygrid = np.linspace(min_i, max_i)

plt.figure(figsize=(8, 6))

# Plot level sets
for c in np.arange(6, 18, 2):
    plt.plot(xgrid, m1(xgrid, ygrid, c), '--k', alpha=0.6)
    
# Plot opposite of feasible set
plt.fill_between(xgrid, c1(xgrid), max_i, color='lightblue')
plt.fill_between(xgrid, c2(xgrid), max_i, color='lightblue')
plt.fill_between(xgrid, min_i, c3(xgrid), color='lightblue')

# Solve for optimum
f = lambda x: 6 - 2/3 * x - (x - 4)
x = root(f, 1).x
y = c3(x)
plt.hlines(y, 0, x, linestyle=':', color='red', label='Optimizer')
plt.vlines(x, 0, y, linestyle=':', color='red')

plt.ylim(min_i, max_i)
plt.xlim(min_i, max_i)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')

# Add a legend
plt.fill_between(xgrid, min_i, min_i, color='white', label='Feasible set')
plt.plot(xgrid, -1 * np.ones(len(xgrid)), '--k', alpha=0.6, label='Level sets')
plt.legend(loc='upper right')

plt.savefig('plot2.png')
plt.show()
```
