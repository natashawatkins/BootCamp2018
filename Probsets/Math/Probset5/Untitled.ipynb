```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
m1 = lambda x, y, c: 5 * x - 4 * y + c
c1 = lambda x: 2/3 * x + 4/3
c2 = lambda x: x/6 - 1/6
c3 = lambda x: 6 - x
```

```python
max_y = 10
max_x = 10
min_y = 0
min_x = 0

xgrid = np.linspace(min_y, max_y)
ygrid = np.linspace(min_x, max_x)

for c in np.arange(0, 10, 0.5):
    plt.plot(xgrid, m1(xgrid, ygrid, c), color='k')
plt.fill_between(xgrid, min_y, c1(xgrid), edgecolor='k', color='grey')
plt.fill_between(xgrid, min_y, c2(xgrid), edgecolor='k', color='grey')
plt.fill_between(xgrid, c3(xgrid), max_y, edgecolor='k', color='grey')
plt.ylim(min_y, max_y)
plt.xlim(min_x, max_x)
plt.show()
```
